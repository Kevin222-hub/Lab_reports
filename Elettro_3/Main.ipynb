```python
import my_lib as my
import matplotlib.pyplot as plt
import numpy as np
import scipy as s
import pandas as pd
from scipy import stats
from math import log10, floor,pi,sqrt,tan

m=0.001
u=0.000001
n=0.000000001
p=0.000000000001
k=1000
M=1000000
G=1000000000
T=1000000000000
```

```python
#Parte 1-2

C = 4.7*n
L = 10*m
Rl = 38.0
R = 453
Rg = 50
```

```python
#Parte 1
D1 = pd.read_csv('Data_1.csv').T.to_numpy() # f[0], Vin[1], Vout[2], dT[3]
fm1 = D1[0]             # Frequenza Misurata
om1 = 2*pi*fm1          # Pulsazione
Hm1 = D1[2]/D1[1]       # Rapporto
sfm1 = 2*pi*D1[3]*u*fm1 # Sfasamento

#Parte 2
D2 = pd.read_csv('Data_2.csv').T.to_numpy()
fm2 = D2[0]
om2 = 2*pi*fm2
Hm2 = D2[2]/D2[1]
```

```python
#1 Resistenza Ampiezza


plt.plot(fm1,Hm1,ls='', marker='.',label='Punti Misurati')

fmin1 = np.min(500)
fmax1 = np.max(100000)
f1 = np.linspace(fmin1,fmax1,100)
o1 = 2*pi*f1
H1 = R/np.sqrt( (R+Rl)**2 + np.power( o1*L - 1/(o1*C),2))
plt.plot(f1,H1,label='Andamento Teorico')


plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Andamento Rapporti Ampiezze Resistore')
plt.ylabel("$|V_{out}/V_{in}|$")
plt.xlabel("$\u03BD\ [Hz]$")
plt.legend(loc="upper left")
```

```python
#1 Zoom Resistenza Ampiezza
plt.plot(fm1,Hm1,ls='', marker='.')#,label='Punti Misurati')
plt.plot(f1,H1)#,label='Andamento Teorico')

Hmax = np.max(Hm1) # Misurato
#Hmax = np.max(H1) # Teorico
H12 = Hmax/sqrt(2)

plt.hlines(H12,fm1[3],fm1[7],ls='--',color='g',label='$\dfrac{V_{out}}{V_{in}} =\dfrac{1}{\sqrt{2}}\ H_{max}$')
plt.vlines(fm1[3],0,H12,ls='--',color='r',label='$\u03BD_{1}=19666\ Hz$') # v_1[3]
plt.vlines(fm1[7],0,H12,ls='--',color='r',label='$\u03BD_{2}=27449\ Hz$') # v_2[7]

plt.vlines(fm1[5],0,np.max(Hm1),ls='--',color='y',label='$\u03BD_{0}=23283\ Hz$') # v_0[5]

plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Zoom Andamento Rapporti Ampiezze Resistore')
plt.ylabel("$|V_{out}/V_{in}|$")
plt.xlabel("$\u03BD\ [Hz]$")
plt.legend(loc="upper left")
plt.xlim(10*k,50*k)
plt.ylim(0.4,1.0)
fm1[5]
```

```python
#2 Resistenza Sfasamento

plt.plot(fm1,sfm1,ls='', marker='.',label='Punti Misurati')

rad1 = -np.arctan( (o1*L - 1/(o1*C) )/(R+Rl) )
plt.plot(f1,rad1,label='Andamento Teorico')

plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Andamento Sfasamento Resistore')
plt.ylabel("$\Delta\Phi\ [rad]$")
plt.xlabel("$\u03BD\ [Hz]$")
plt.legend(loc="lower left")
```

```python
#2 Zoom Resistenza Sfasamento

plt.plot(fm1,sfm1,ls='', marker='.')#,label='Punti Misurati')
plt.plot(f1,rad1)#,label='Andamento Teorico')


SF0 = pi/4

plt.hlines(SF0,fmin1,fm1[3],ls='--',color='g',label='$\Delta\Phi\ =\pm\dfrac{\pi}{4}$') # v_1[3]
plt.vlines(fm1[3],-1.5,sfm1[3],ls='--',color='r',label='$\u03BD_{1}=19666\ Hz$')

plt.hlines(-SF0,fmin1,fm1[7],ls='--',color='g') # v_2[7]
plt.vlines(fm1[7],-1.5,sfm1[7],ls='--',color='r',label='$\u03BD_{2}=27449\ Hz$')

plt.vlines(fm1[5],-1.5,0,ls='--',color='y',label='$\u03BD_{0}=23283\ Hz$') # v_0[5]

plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Zoom Andamento Sfasamento Resistore')
plt.ylabel("$\Delta\Phi\ [rad]$")
plt.xlabel("$\u03BD\ [Hz]$")
plt.xlim(10*k + 5000 ,35*k)
plt.legend(loc="upper right")
```

```python
#1 Condensatore Ampiezza
plt.plot(fm2,Hm2,ls='', marker='.',label='Punti Misurati')

fmin2 = np.min(500)
fmax2 = np.max(100000)
f2 = np.linspace(fmin2,fmax2,100)
o2 = 2*pi*f2
H2 = (1/(o2*C)) / np.sqrt( (R+Rl)**2 + np.power( o2*L - 1/(o2*C),2))
plt.plot(f2,H2,label='Andamento Teorico')


plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Andamento Rapporti Ampiezze Condensatore')
plt.ylabel("$|V_{out}/V_{in}|$")
plt.xlabel("$\u03BD\ [Hz]$")
plt.legend(loc="upper left")
```
