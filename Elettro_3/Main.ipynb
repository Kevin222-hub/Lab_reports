```python
import my_lib as my
import matplotlib.pyplot as plt
import numpy as np
import scipy as s
import pandas as pd
from scipy import stats
from math import log10, floor,pi,sqrt,tan

m=0.001
u=0.000001
n=0.000000001
p=0.000000000001
k=1000
M=1000000
G=1000000000
T=1000000000000
```

```python
#Parte 1-2

C = 4.7*n
L = 10*m
Rl = 38.0
R = 453
Rg = 50
```

```python
#Parte 1
D1 = pd.read_csv('Data_1.csv').T.to_numpy() # f[0], Vin[1], Vout[2], dT[3]
fm1 = D1[0]             # Frequenza Misurata
om1 = 2*pi*fm1          # Pulsazione
Hm1 = D1[2]/D1[1]       # Rapporto
sfm1 = 2*pi*D1[3]*u*fm1 # Sfasamento

#Parte 2
D2 = pd.read_csv('Data_2.csv').T.to_numpy()
fm2 = D2[0]
om2 = 2*pi*fm2
Hm2 = D2[2]/D2[1]
```

```python
#1 Resistenza Ampiezza


plt.plot(fm1,Hm1,ls='', marker='.',label='Punti Misurati')

fmin1 = np.min(500)
fmax1 = np.max(100000)
f1 = np.linspace(fmin1,fmax1,100)
o1 = 2*pi*f1
H1 = R/np.sqrt( (R+Rl)**2 + np.power( o1*L - 1/(o1*C),2))
plt.plot(f1,H1,label='Andamento Teorico')


H0 = np.max(H1)/sqrt(2)
#om3 = 7993
plt.hlines(H0,fmin1,fmax1,ls='--',color='g',label='$\dfrac{V}{V_0} =\dfrac{V_{MAX}}{V_0\sqrt{2}}}$')
#plt.vlines(om3,0,H0,ls='--',color='r',label='$\u03BD_{0}=7993\ Hz$')


plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Andamento Tensione')
plt.ylabel("$|V/V_0|$")
plt.xlabel("$\u03BD[Hz]$")
plt.legend(loc="upper left")
```

```python
#2 Resistenza Sfasamento

plt.plot(fm1,sfm1,ls='', marker='.',label='Punti Misurati')

rad1 = -np.arctan( (o1*L - 1/(o1*C) )/(R+Rl) )
plt.plot(f1,rad1,label='Andamento Teorico')


SF0 = pi/4
#om3 = 8980
plt.hlines(SF0,fmin1,fmax1,ls='--',color='g',label='$\Delta\Phi\ =\pm\dfrac{\pi}{4}$')
plt.hlines(-SF0,fmin1,fmax1,ls='--',color='g')
#plt.vlines(om3,SF3,-1.5,ls='--',color='r',label='$\u03BD_{0}=8980\ Hz$')


plt.xscale("log")
plt.grid(which="both", ls="--",color='0.7')
plt.title('Andamento Sfasamento')
plt.ylabel("$\Delta\Phi\ [rad]$")
plt.xlabel("$\u03BD[Hz]$")
plt.legend(loc="lower left")
```
