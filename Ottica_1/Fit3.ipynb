```python
import my_lib as my
import matplotlib.pyplot as plt
import numpy as np
import scipy as s
import pandas as pd
from scipy import stats
from math import log10, floor,pi,sqrt,tan

m=0.001
u=0.000001
n=0.000000001
p=0.000000000001
k=1000
M=1000000
G=1000000000
T=1000000000000

#funzioni
#Pt 1
def Lambda(x,Z,F,m): # nm
    L = x*F/(Z*m) *M  # M: normalizzazione L[nm]
    return L

#errore
def SigmaLambda(x,sx,Z,sZ,F,sF,m):
    dx = Lambda(sx,Z, F,m)
    dF = Lambda( x,Z,sF,m)
    dZ = F*x/(np.power(Z,2)*m)*sZ *M
    sL = np.sqrt(np.power(dx,2) + np.power(dZ,2) + np.power(dF,2) )
    return sL
```

```python
Z = 852     # mm 
sZ  = 2

df1   = pd.read_csv('data1.csv')
ArrX = np.asarray(df1.iloc[:,8:12])
ArrY = np.asarray(df1.iloc[:,3:7])
ArrsY = ArrY*0.05

Array_X = np.asarray([])                        
Array_sX = np.asarray([])
Array_L = np.asarray([])                         
Array_sL = np.asarray([])
#df1
```

```python
#0 - 0.5

T = 0
m  =1
F  =0.2
sF =0.001
MAX = 10.9

def Seno(x,a,b,c):
    return a*(x-b)**2 + c

X  = ArrX[T]
Y  = ArrY[T]
sY = ArrsY[T]
    
x = np.linspace(np.min(X),np.max(X),100)
[a,b,c],pcov = s.optimize.curve_fit(Seno,X,Y,sigma = sY,absolute_sigma = True,method = 'trf',bounds = ([0,0,30],[100,100,60]))
[sa,sb,sc] = np.sqrt(np.diag(pcov))
plt.errorbar(X,Y,sY, ls='', marker='.',label = '{}'.format(T))
plt.plot(x,Seno(x,a,b,c))
plt.grid()
plt.legend()
plt.show()

Value =  np.abs(b - MAX) - 0.5
sValue = sb

L = Lambda(Value,Z,F,m)
sL = SigmaLambda(Value,sValue,Z,sZ,F,sF,m)
Array_X = np.append(Array_X,Value)                         
Array_sX = np.append(Array_sX,sValue)
Array_L = np.append(Array_L,L)                         
Array_sL = np.append(Array_sL,sL)

print('N°',T)
print('Xmin = ',Value,'+-',sValue)
print('L =',L,'+-',sL)
#print('a = ',a,'+-',sa)
#print('b = ',b,'+-',sb)
```
